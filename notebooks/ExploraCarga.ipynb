```python
FIXED_WIDTH = '/home/ivan/pybr/AJNA_MOD/sentinela/files/P1/'
```

```python
import glob
sch_files = sorted([file for file in glob.glob(FIXED_WIDTH+'*.sch')])
txt_files = sorted([file for file in glob.glob(FIXED_WIDTH+'*.txt')])
```

```python
ENCODE = 'latin1'
path = FIXED_WIDTH
mask_txt='0.txt'

### Igual csv_handler
sch_name = sch_files[7]
txt_name = glob.glob(os.path.join(
    path, '*' + os.path.basename(sch_name)[3:-4] + mask_txt))[0]
with open(sch_name, encoding=ENCODE,
          newline='') as sch_file, \
        open(txt_name, encoding=ENCODE,
             newline='') as txt_file:
    sch_content = sch_file.readlines()
    reader = csv.reader(txt_file, delimiter='\t')
    txt_content = [linha for linha in reader]

        
print(sch_content)
sch = sch_content
cabecalhos = []
larguras = []
for ind in range(len(sch)):
    if not isinstance(sch[ind], str):
        sch[ind] = str(sch[ind], ENCODE)
    linha = sch[ind]
    position_equal = linha.find('="')
    position_quote = linha.find('" ')
    position_col = linha.find('Col')
    if position_equal != -1 and position_col == 0:
        cabecalhos.append(linha[position_equal + 2:position_quote])
        position_width = linha.find('Width')
        larguras.append(linha[position_width + 6:-2])
print(cabecalhos)
print(larguras)
```

```python
dfs = {}
for r in range(len(files)):
    print(files[r])
    with open(os.path.join(CARGA_BASE, files[r]), 'r',
              encoding='latin1', newline='') as arq:
        reader = csv.reader(arq)
        lista = [linha for linha in reader]
    width_header = len(lista[0])
    for ind, linha in enumerate(lista):
        width_linha = len(linha)
        if width_linha != width_header:
            print('Erro: ', 'Header ', width_header, 'Linha ', width_linha)
            print(lista[0])
            print(lista[ind-1])
            print(linha)
            for header, certo, errado in zip(lista[0], lista[ind-1], lista[ind]):
                print(header, 'Certo', certo[:10], 'Errado', errado[:10])
        while width_linha > width_header:
            # Caso haja colunas "sobrando" na linha, retirar
            # uma coluna nula
            for index, col in enumerate(linha):
                if isinstance(col, str) and not col:
                    print('****Primeira coluna nula:', index)
                    width_linha -= 1
                    linha.pop(index)
                    break

        
    column_lenght = len(lista[1])
    header_lenght = len(lista[0])
    lenght = header_lenght
    if column_lenght < header_lenght:
        lenght = column_lenght
    print('Headers', header_lenght, 'Columns', column_lenght, 'Final', lenght)
    df = pd.DataFrame(lista[1:], columns=lista[0])
    dfs[files[r]] = df
```
